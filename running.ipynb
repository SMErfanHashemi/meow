```python
import json

# ‚îÄ‚îÄ‚îÄ CONFIG ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
VEST_CLASS    = "a vest on"
NOVEST_CLASS  = "a novest"
HELM_CLASS    = "a helmet on"
NOHELM_CLASS  = "a nohelmet"
IOU_THRESH    = 0.1    # vest IoU threshold
MARGIN        = 20     # px to expand helmet/no-helmet boxes
# ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ

def iou(boxA, boxB):
    xA, yA = max(boxA[0], boxB[0]), max(boxA[1], boxB[1])
    xB, yB = min(boxA[2], boxB[2]), min(boxA[3], boxB[3])
    inter = max(0, xB - xA) * max(0, yB - yA)
    areaA = (boxA[2] - boxA[0]) * (boxA[3] - boxA[1])
    areaB = (boxB[2] - boxB[0]) * (boxB[3] - boxB[1])
    return inter / (areaA + areaB - inter) if (areaA + areaB - inter) else 0

def expand(box, m):
    x1, y1, x2, y2 = box
    return [x1-m, y1-m, x2+m, y2+m]

def overlaps(a, b):
    return max(0, min(a[2],b[2]) - max(a[0],b[0])) * \
           max(0, min(a[3],b[3]) - max(a[1],b[1])) > 0

# Load once
with open(r'results\result.json','r') as f:
    data = json.load(f)
person_counter = 1
output = {}

for img, regions in data.items():
    # 1) Separate original detections
    persons = [r for r in regions if r["cap"] == "a person"]
    vests   = [r for r in regions if r["cap"] == VEST_CLASS]
    novests = [r for r in regions if r["cap"] == NOVEST_CLASS]
    helms   = [r for r in regions if r["cap"] == HELM_CLASS]
    nohelms = [r for r in regions if r["cap"] == NOHELM_CLASS]

    # 2) Helpers
    vest_iou_fn   = lambda box: max((iou(box, v["box"]) for v in vests), default=0)
    novest_iou_fn = lambda box: max((iou(box, n["box"]) for n in novests), default=0)
    helms_exp     = [expand(h["box"], MARGIN) for h in helms]
    nohelms_exp   = [expand(nh["box"], MARGIN) for nh in nohelms]
    person_boxes  = [p["box"] for p in persons]

    # 3) Label persons
    person_regions = []
    for p in persons:
        pb = p["box"]

        # vest status
        vi  = vest_iou_fn(pb)
        nvi = novest_iou_fn(pb)
        if vi > IOU_THRESH:
            vest_stat = "with vest"
        elif nvi > IOU_THRESH:
            vest_stat = "without vest"
        else:
            vest_stat = "undetected vest"

        # helmet status
        if any(overlaps(pb, he) for he in helms_exp):
            helm_stat = "with helmet"
        elif any(overlaps(pb, nh) for nh in nohelms_exp):
            helm_stat = "without helmet"
        else:
            helm_stat = "undetected helmet"

        person_regions.append({
            "box":   pb,
            "score": p["score"],
            "cap":   f"person{person_counter} {vest_stat} and {helm_stat}"
        })
        person_counter += 1

    # 4) Prune used vest/helmet boxes
    leftovers = []
    for r in regions:
        c, b = r["cap"], r["box"]
        if c in {VEST_CLASS, NOVEST_CLASS} and any(iou(b, pb) > 0 for pb in person_boxes):
            continue
        if c in {HELM_CLASS, NOHELM_CLASS} and any(overlaps(b, pb) for pb in person_boxes):
            continue
        if c != "a person":
            leftovers.append(r)

    # 5) Post-process leftovers
    caps = {r["cap"] for r in leftovers}
    if NOHELM_CLASS in caps and NOVEST_CLASS in caps:
        leftovers = [r for r in leftovers if r["cap"] not in {NOHELM_CLASS, NOVEST_CLASS}]
    if HELM_CLASS in caps and VEST_CLASS in caps:
        new_left = []
        for r in leftovers:
            if r["cap"] == HELM_CLASS:
                new_left.append({**r, "cap": "helmet without user"})
            elif r["cap"] == VEST_CLASS:
                new_left.append({**r, "cap": "vest without user"})
            else:
                new_left.append(r)
        leftovers = new_left

    final_left = [
        r for r in leftovers
        if r["cap"] not in {HELM_CLASS, NOHELM_CLASS, VEST_CLASS, NOVEST_CLASS}
    ]

    output[img] = person_regions + final_left

# Save
with open(r'results\result_postprocessed.json', 'w') as f:
    json.dump(output, f, indent=2)

print("‚úÖ Done. Saved to results\\result_postprocessed.json ‚Äî no raw PPE boxes remain.")
```
